```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
```

```python
dataset = pd.read_excel("golf_df.xlsx")
dataset.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(dataset.temp)
dataset.temp = le.transform(dataset.temp)

le.fit(dataset.clima)
dataset.clima = le.transform(dataset.clima)

le.fit(dataset.umidade)
dataset.umidade = le.transform(dataset.umidade)

le.fit(dataset.vento)
dataset.vento = le.transform(dataset.vento)

le.fit(dataset.jogar)
dataset.jogar = le.transform(dataset.jogar)

dataset.head()
```

```python
X = dataset.iloc[:, :-1]
y = dataset[["jogar"]]
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
predict = model.predict(X)
```

```python
X["prediction"] = predict
y = pd.DataFrame(y)
X["real"] = y["jogar"]
X.head(20)
```
